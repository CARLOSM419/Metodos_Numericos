# Ejercicio 2 : Error Numérico de una aproximación
Objetivo : calcular el error al aproximar raiz cuadrada

03/02/2026

```python
import math  # Proporciona funciones y constantes matemáticas avanzadas para trabajar con números reales
real = math.sqrt(2)#Calcula la raíz cuadrada de 2 usando la función sqrt() y la guarda en la variable "real"
aprox = 1.41# Guarda en la variable "aprox" un valor aproximado de la raíz de 2
# Calcula el error absoluto restando el valor aproximado al real y aplicando valor absoluto
error_absoluto = abs(real - aprox)
error_relativo = error_absoluto / real# Calcula el error relativo dividiendo el error absoluto entre el valor real
print(error_absoluto, error_relativo)# Imprime en pantalla el error absoluto y el error relativo
```
