# Ejercicio 3 : Optimización numérica simple.
Problema : Minizar f(x)=(x-3)**2

```python
x = 0# Inicializa la variable x en 0 (punto de partida)
alpha = 0.1# Define la tasa de aprendizaje (learning rate), que controla el tamaño del paso en cada iteración
 # Ejecuta el ciclo 20 veces (20 iteraciones del algoritmo)
for _ in range(20):
  grad = 2 * (x - 3) # Calcula el gradiente (derivada) de la función f(x) = (x - 3)^2
  x = x - alpha * grad# Actualiza el valor de x moviéndolo en dirección opuesta al gradiente
print(x)# Imprime el valor final de x después de 20 iteraciones
```
