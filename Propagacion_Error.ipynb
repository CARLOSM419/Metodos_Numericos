# Ejercicio 3 : Propagación del error

17/02/2026

```python
x = 2# Define el valor de la variable independiente x
dx = 0.01 # Define un pequeño error o variación en x (diferencial)
 # Calcula el valor de la función y = x^2
y = x**2
dy = abs (2*x) * dx# Calcula el error aproximado en y usando diferenciales
# Imprime el valor de y
print("y : ",y)
print("Error aproximado en y : ",dy)# Imprime el error aproximado propagado en y
```
