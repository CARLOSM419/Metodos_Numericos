# Sesgo


```python
import numpy as np# Importa la librería NumPy para trabajar con arreglos y generar números aleatorios
# Genera 1000 números aleatorios con distribución normal
datos_reales = np.random.normal(0,1,1000)
# Define un valor de sesgo (bias) que se agregará a los datos
sesgo = 0.5
print(datos_reales)# Imprime los datos originales generados aleatoriamente

# Suma 0.5 a cada dato, desplazando toda la distribución hacia la derecha
datos_sesgados = datos_reales + sesgo
print(datos_sesgados)# Imprime los datos después de aplicar el sesgo
print("Media real", np.mean(datos_reales))# Calcula e imprime la media de los datos originales (debe estar cerca de 0)
print("Media Sesgada", np.mean(datos_sesgados))# Calcula e imprime la media de los datos con sesgo (debe estar cerca de 0.5)
```
