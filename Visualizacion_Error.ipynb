```python
import numpy as np# Importa NumPy para trabajar con arreglos numéricos
import matplotlib.pyplot as plt# Importa matplotlib para crear gráficas
# Genera 100 valores igualmente espaciados entre -1 y 5
x = np.linspace(-1,5,100)
y = (x-2)**2# Calcula la función y = (x - 2)^2 para cada valor de x
# Dibuja la gráfica de la función
plt.plot(x,y)
plt.xlabel("x")# Etiqueta el eje horizontal
plt.ylabel("y")# Etiqueta el eje vertical
plt.title("Funcion de costo")# Agrega un título a la gráfica
plt.show()# Muestra la gráfica en pantalla
```
