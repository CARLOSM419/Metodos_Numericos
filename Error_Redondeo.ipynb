# Ejercicio 1 : Error de redondeo
comparaciones directas pueden fallar

en IA se usan tolerancias

21/02/2026

```python
a = 0.1 + 0.2 # Suma dos números decimales en punto flotante
b = 0.3 # Asigna el valor 0.3 a la variable b

 # Imprime el resultado de la suma (verás algo como 0.30000000000000004)
print(a)
print(a == b )# Compara si a es exactamente igual a b (normalmente dará False por precisión en punto flotante)
print(abs(a - b))# Calcula el valor absoluto de la diferencia entre a y b (muestra el pequeño error numérico)
```
