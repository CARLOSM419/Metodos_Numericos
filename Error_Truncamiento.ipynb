```python
import math# Importa la librería math para usar funciones matemáticas como la exponencial
 # Define el valor de x
x= 1
aprox = 1 + x + (x**2)/2# Calcula una aproximación de e^x usando los primeros términos
real = math.exp(x)
# Calcula el valor real de e^x usando la función exp() de la librería math
error = abs(real - aprox)# Calcula el error absoluto entre el valor real y la aproximación
# Imprime el valor aproximado
print("Aprox : ",aprox)
print("Real : ",real)# Imprime el valor real calculado con math.exp()
print("Error : ",error)# Imprime el error absoluto
```
