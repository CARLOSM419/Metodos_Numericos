```python
import numpy as np# Importa la librería NumPy para trabajar con matrices y álgebra lineal
# Define la matriz de coeficientes del sistema de ecuaciones
A = np.array([[3,2],
              [1,2]])
# Define el vector de términos independientes
b = np.array([5,5])
# Resuelve el sistema de ecuaciones lineales A·x = b
# usando métodos de álgebra lineal
x = np.linalg.solve(A,b)
print(x)# Imprime el vector solución del sistema
```
