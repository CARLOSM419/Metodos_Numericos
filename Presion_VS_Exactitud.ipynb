```python
import numpy as np# Importa la librería NumPy para trabajar con arreglos y operaciones numéricas
mediciones = np.array([9.8,9.81,9.79,9.80])# Crea un arreglo (array) con varias mediciones experimentales
# Define el valor real o teórico con el que se compararán las mediciones
valor_real = 9.81
# Calcula la media (promedio) de las mediciones usando la función mean()
media = np.mean(mediciones)
# Calcula el error absoluto restando la media al valor real y aplicando valor absoluto
error = abs(media - valor_real)
# Imprime el valor promedio calculado
print("La media es: ", media)
print("El error absoluto es: ", error)# Imprime el error absoluto entre la media y el valor real
```
