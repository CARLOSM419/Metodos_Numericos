```python
from scipy.optimize import minimize# Esta función sirve para encontrar el mínimo de una función
# Define una función anónima (lambda)
f = lambda x : (x-2)**2
# Llama a la función minimize para encontrar el mínimo de f
resultado  = minimize(f, x0=0)# Imprime el valor de x donde la función alcanza su mínimo
print(resultado.x)
```
